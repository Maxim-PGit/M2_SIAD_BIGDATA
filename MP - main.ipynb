### Différenciation entre vacances locales, régionales et nationales

```python
#Pour process le fichier de test, modifier le chemin de "train = ", décommenter la ligne suivante et modifier le chemin décriture csv plus bas
holidays_events = pd.read_csv("data/initial/holidays_events.csv")
train = pd.read_csv("data/initial/train.csv")
#train['sales'] = np.nan
stores = pd.read_csv("data/initial/stores.csv")
```

```python
#holidays_events
holidays_events = holidays_events[holidays_events.transferred != True]
holidays_events = holidays_events.drop(['type','description','transferred'], axis=1)
holidays_events['datePlaceKey'] = holidays_events['date'].astype(str) + '_' + holidays_events['locale_name']

```

```python
display(stores)
```

```python

    #Train et emplacements magasin
stores = stores.drop(['type','cluster'], axis=1)
trainStores = train.merge(stores, left_on='store_nbr', right_on='store_nbr')

#Merge
    #Niveau local
holidaysLocal = holidays_events[holidays_events.locale == 'Local']
holidaysLocal = holidaysLocal.drop(['locale','date','locale_name'], axis=1)
    #Niveau Regional
holidaysRegional = holidays_events[holidays_events.locale == 'Regional']
holidaysRegional = holidaysRegional.drop(['locale','date','locale_name'], axis=1)
    #Niveau National
holidaysNational = holidays_events[holidays_events.locale == 'National']
holidaysNational = holidaysNational.drop(['locale','datePlaceKey','locale_name'], axis=1)
```

```python
display(trainStores)
```

```python

    #Création ID (magasin lié à dateVille et lié à dateRégion)
trainStores['dateCityKey'] = trainStores['date'].astype(str) + '_' + trainStores['city']
trainStores['dateStateKey'] = trainStores['date'].astype(str) + '_' + trainStores['state']

    #Niveau local
trainStoreHolidaysCity   = trainStores.merge(holidaysLocal, left_on='dateCityKey', right_on='datePlaceKey')
trainStoreHolidaysCity = trainStoreHolidaysCity.drop(['id','date','datePlaceKey','dateStateKey','store_nbr','family','sales','onpromotion','city','state'], axis=1)
trainStoreHolidaysCity['flagHolidayCity'] = 1

    #Niveau Regional
trainStoreHolidaysRegion = trainStores.merge(holidaysRegional, left_on='dateStateKey', right_on='datePlaceKey')
trainStoreHolidaysRegion = trainStoreHolidaysRegion.drop(['id','date','datePlaceKey','dateCityKey','store_nbr','family','sales','onpromotion','city','state'], axis=1)
trainStoreHolidaysRegion['flagHolidayRegion'] = 1

    #Niveau National
trainStoreHolidaysGlobal = trainStores.merge(holidaysNational, left_on='date', right_on='date')
trainStoreHolidaysGlobal = trainStoreHolidaysGlobal.drop(['id','dateCityKey','dateStateKey','store_nbr','family','sales','onpromotion','city','state'], axis=1)
trainStoreHolidaysGlobal['flagHolidayGlobal'] = 1
```

```python
trainStoreHolidaysCity = trainStoreHolidaysCity.drop_duplicates()
trainStoreHolidaysRegion = trainStoreHolidaysRegion.drop_duplicates()
trainStoreHolidaysGlobal = trainStoreHolidaysGlobal.drop_duplicates()

display(trainStoreHolidaysCity)
display(trainStoreHolidaysRegion)
display(trainStoreHolidaysGlobal)
```

# Fichier Oil

```python
oil = pd.read_csv("data/initial/oil.csv")
```

### Recherche de valeurs manquantes

```python
#Gestion du cas où les valeurs manquantes sont au format texte (nc, mq, nan etc)
colName = "dcoilwtico"
oil[colName] = pd.to_numeric(oil[colName], errors='coerce')
print("Nombre de valeurs manquantes pour '" + colName + "' : " + str(oil[colName].isnull().sum()))
display(oil)
```

```python
# Convertir la colonne 'date' en type datetime
oil['date'] = pd.to_datetime(oil['date'])

# Trier le DataFrame par la colonne 'date'
oil = oil.sort_values(by='date')

# Remplacer les valeurs manquantes par la moyenne entre le jour précédent et le jour suivant
oil['dcoilwtico'] = oil['dcoilwtico'].fillna((oil['dcoilwtico'].shift() + oil['dcoilwtico'].shift(-1)) / 2)

# Trier le DataFrame par la colonne 'date' d'origine
oil = oil.sort_values(by='date')

# Afficher le DataFrame résultant
print(oil)

# on regarde s'il y a encore des valeurs manquante
oil[oil['dcoilwtico'].isnull()]

# Afficher les lignes sélectionnées
#print(ligne_1173_et_suivantes)

#création d'une séquence pour ajouter les week-end
oil['date'] = pd.to_datetime(oil['date'])
dates_attendues = pd.date_range(start='2013-01-01', end='2017-08-31', freq='D')

# Trouver les dates manquantes
dates_manquantes = dates_attendues[~dates_attendues.isin(oil['date'])]

# Afficher les dates manquantes (si aucune date n'est manquante, la liste sera vide)
print("Dates manquantes :", dates_manquantes)
liste_date_manquante = len(dates_manquantes)
print('il y a', liste_date_manquante, 'dates qui manquent')

# Créer un DataFrame avec les dates manquantes
df_dates_manquantes = pd.DataFrame({'date': dates_manquantes, 'dcoilwtico': None}) 

# Fusionner les DataFrames pour ajouter les dates manquantes
oil = pd.concat([oil, df_dates_manquantes])

# Trier le DataFrame par la colonne des dates
oil = oil.sort_values(by='date')

# Réindexer le DataFrame
oil = oil.reset_index(drop=True)

# Afficher le DataFrame complet avec les dates manquantes ajoutées
print(oil)

# Trier le DataFrame par la colonne 'date'
oil = oil.sort_values(by='date')

# Interpoler linéairement les valeurs manquantes dans la colonne 'dcoilwtico'
oil['dcoilwtico'] = oil['dcoilwtico'].interpolate()

# Afficher le DataFrame résultant
print(oil)

# Imputation de la 1ère valeur
oil[oil['dcoilwtico'].isnull()]
oil.at[0, 'dcoilwtico'] = 93.12

print(oil)
```
