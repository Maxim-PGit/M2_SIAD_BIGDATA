```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib

#from ydata_profiling import ProfileReport
```

```python
oil = pd.read_csv("data/oil.csv")
holidays_events = pd.read_csv("data/holidays_events.csv")
sample_submission = pd.read_csv("data/sample_submission.csv")
stores = pd.read_csv("data/stores.csv")
test = pd.read_csv("data/test.csv")
train = pd.read_csv("data/train.csv")
```

```python
# holidays_events

# Reprsentation graphiques de nos variables

col_categorielles = ['type','transferred','locale']

for column in col_categorielles:
    
    # Crer le diagramme en barres avec pourcentage
    
    plt.figure(figsize=(12, 6))
    ax = sns.countplot(x=column, data=holidays_events , palette='viridis')
    total = len(holidays_events[column])

    # Ajouter les pourcentages sur les barres
    
    for p in ax.patches:
        pct = '{:.1f}%'.format(100 * p.get_height() / total)
        x = p.get_x() + p.get_width() / 2 - 0.05
        y = p.get_height() + 0.3
        ax.annotate(pct, (x, y), ha='center', fontsize=10, color='black')

    plt.title(f'Distribution de {column}')
    plt.xticks(rotation=45, ha='right')

```
